```python
# This one is in-built in Seaborn
# Load the Iris dataset 
data = sns.load_dataset("iris")
```

```python
# 1. Explore the data
print("=== First 5 Rows ===")
print(data.head())

print("\n=== Info ===")
print(data.info())

print("\n=== Summary Statistics ===")
print(data.describe())
```

```python
data.info()
```

```python
# 2. Calculate basic statistics
print("\n=== Mean of Numerical Features ===")
print(data.mean(numeric_only=True))

print("\n=== Median of Numerical Features ===")
print(data.median(numeric_only=True))

print("\n=== Mode of Numerical Features ===")
print(data.mode().iloc[0])  # Mode may have multiple rows, take first

print("\n=== Standard Deviation of Numerical Features ===")
print(data.std(numeric_only=True))

print("\n=== Correlation Matrix ===")
print(data.corr(numeric_only=True))
```

```python
# ===== 1. Basic Line Plot =====
plt.figure(figsize=(8, 4))
plt.plot(data.index, data['sepal_length'], label='Sepal Length', marker='o')
plt.title("Line Plot of Sepal Length")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.legend()
plt.grid()
plt.show()
```

```python
# ===== 2. Scatter Plot =====
plt.figure(figsize=(8, 4))
plt.scatter(data['sepal_length'], data['petal_length'], c='green', alpha=0.6)
plt.title("Scatter Plot: Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.grid()
plt.show()
```

```python
# ===== 3. Histogram =====
plt.figure(figsize=(8, 4))
sns.histplot(data['sepal_width'], bins=15, kde=True, color='purple')
plt.title("Histogram of Sepal Width")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# ===== 4. Box Plot =====
plt.figure(figsize=(8, 4))
sns.boxplot(x='species', y='petal_length', data=data, palette='Set2')
plt.title("Box Plot of Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.show()
```
